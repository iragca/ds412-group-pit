````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('https://raw.githubusercontent.com/iragca/iragca/refs/heads/main/configs/matplotlib/styles/cmr10.mplstyle')

def load_data():
    if IS_COLAB:
        return pd.read_csv('https://raw.githubusercontent.com/iragca/ds412-group-pit/refs/heads/master/data/external/milk_.csv')
    else:
        from lib.config import Directories
        return pd.read_csv(Directories.EXTERNAL_DATA_DIR.value / 'milk_.csv')

data = load_data()
````

````python
# Preprocess the data
processed_data = data.copy()
processed_data['index'] = pd.to_datetime(data['index'], format='%b %Y')
processed_data.rename(columns={'value': 'Milk Production (pounds)', 'index': 'Date'}, inplace=True)


# Plot the data
sns.lineplot(data=processed_data, x='Date', y='Milk Production (pounds)')
plt.title('Monthly Milk Production Over Time')
````
